```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('swiggysales.csv')
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
platform_counts = df['Platform'].value_counts()
platform_counts
```

```python
plt.figure(figsize=(8,6))
plt.barh(platform_counts.index, platform_counts.values, color='pink',edgecolor='black')
plt.xlabel('Platform')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Platform')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
plt.pie(platform_counts.values, labels=platform_counts.index, autopct='%1.1f%%',explode=(0.1,0,0), startangle=140,wedgeprops={
    'edgecolor':'black'
})
plt.title('Distribution of Orders by Platform')
plt.show()
```

```python
product_counts=df['Product Category'].value_counts()
product_counts
```

```python
plt.pie(product_counts.values,labels=product_counts.index,autopct='%1.1f%%',wedgeprops={
    'edgecolor':'black'
},startangle=140)

plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# suppose we want to see top 5 customers accoring to the given ordres
top_customers=df['Customer ID'].value_counts().head(5)
top_customers
```

```python
plt.figure(figsize=(8,6))
plt.bar(top_customers.index, top_customers.values, color='orange',edgecolor='black')
plt.xlabel('Customer ID')
plt.ylabel('Number of Orders')
plt.title('Top 5 Customers by Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()  
```

```python
# i want to check the refund or not refund accorss the different platforms
refund_counts=df.groupby(['Platform','Refund Requested']).size().unstack()
refund_counts.plot(kind='bar', stacked=True, figsize=(8,6), color=['lightgreen','salmon'],edgecolor='black')
plt.xlabel('Platform')
plt.ylabel('Number of Orders')
plt.title('Refund Status by Platform')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
