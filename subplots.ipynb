```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,25,15,30]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Chart')


plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('BarChart')


plt.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,25,15,30]
# fig,ax=plt.subplots(nrows,ncols,figsize(width,height))

ax[0].plot(x,y)
ax[0].set_title('Line plot')

ax[1].bar(x,y,color='green')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparision of multiple charts')

plt.tight_layout()
plt.show()
```
