HOW CAN WE COMPARE TOP 3 PLATFORMS THE PUBLISHERS USED

```python
df.head()
```

```python
top_5=df[['Publisher']].value_counts().head(5)
top_5
```

```python
# what are the top-3 platforms the publisher used
plt.figure(figsize=(12,8))
sns.countplot(data=df,x='Publisher',hue='Platform')
plt.show()
```

```python
# Top 5 publisher
top_pub=df['Publisher'].value_counts().index[:5]
```

```python
# Top 5 Genre
top_gen=df['Genre'].value_counts().index[:5]
top_gen
```

```python
# Top 5 Platform
top_plat=df['Platform'].value_counts().index[:5]
top_plat
```

```python
top5data=df.loc[(df['Publisher'].isin(top_pub) & df['Genre'].isin(top_gen) & df['Platform'].isin(top_plat))]
top5data
```

```python
plt.figure(figsize=(8, 8))
sns.countplot(data=top5data,x='Publisher',hue='Platform') # hue means segregation
plt.xticks(rotation=45)
```
