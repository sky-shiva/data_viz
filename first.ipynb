```python
df.head()
```

```python
# categorical columns
df['Genre'].unique()
```

```python
# value_counts give the count of columns in each genre
count_genre=df['Genre'].value_counts()
count_genre
```

I can find top 5 genres


```python
# we can use barchart or the count chart(sns) to visualize the count of each genre

#BAR CHART

x_bar=count_genre.index
y_bar=count_genre.values
x_bar
```

```python
y_bar
```

```python
#plt.bar(x_bar,y_bar)
plt.figure(figsize=(10,8))
plt.bar(x_bar,y_bar)
```

```python
# how to rotate x axis labels
plt.figure(figsize=(11,5))
plt.bar(x_bar,y_bar,color='purple',edgecolor='black')
plt.xticks(rotation=90)
plt.title('count of each genre',fontsize=20,color='black')
plt.xlabel('Genre',color='red')
plt.ylabel('Count',color='red')
plt.show()
```

```python
# same plot through seaborn
sns.countplot(data=df,x='Genre',order=df['Genre'].value_counts().index,color='blue',edgecolor='black')
```

PIE CHART


```python
df
```

```python
sales_data=df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
sales_data
```

```python
region_sales=sales_data.T.sum(axis='columns')
region_sales
```

```python
# if want percentage / proportion we will use pie plot

plt.pie(region_sales,labels=region_sales.index,startangle=90,explode=(0.2,0,0,0),autopct='%1.2f%%',wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})
plt.title("Total Sales Across Various Regions")
plt.show()
```
