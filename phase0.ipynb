```python
import matplotlib.pyplot as plt
```

```python
# line plot

x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.xlabel('Number of hours Studied by student',color='Blue')
plt.ylabel('Marks Scored')
plt.legend()
plt.grid()
plt.show()
```

```python
x=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
y=[23,6,7,8,9,8,1]
z=[1,2,3,4,5,6,7]
plt.figure(figsize=(8,5))  # this /should be done before the plot
plt.plot(x,y,color='b',linewidth=3,label='Cakes sales')
plt.plot(x,z,label='X-z plot',linewidth=3,color='black')
plt.title('Sales of Bakery',color='Black')
plt.xlabel('Days')
plt.ylabel('No of cakes sold',color='Black')
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.xlim(0,6) #only show x values from 0 to 6
plt.ylim(0,) #only show y values from 0 to 6
plt.grid()
```

```python
# sales data
product=['Laptop','Apple','Samsung','Samsung Fold']
earned=[120000,89000,120000,180000]

plt.barh(product,earned,color='orange',label='Products')
plt.xlabel('Product Name')
plt.ylabel('Total Earnings')
plt.title('Total Earnings To Products')
plt.legend()
```
