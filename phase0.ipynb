```python
import matplotlib.pyplot as plt
```

```python
# line plot

x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.xlabel('Number of hours Studied by student',color='Blue')
plt.ylabel('Marks Scored')
plt.legend()
plt.grid()
plt.show()
```

```python
x=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
y=[23,6,7,8,9,8,1]
z=[1,2,3,4,5,6,7]
plt.figure(figsize=(8,5))  # this /should be done before the plot
plt.plot(x,y,color='b',linewidth=3,label='Cakes sales')
plt.plot(x,z,label='X-z plot',linewidth=3,color='black')
plt.title('Sales of Bakery',color='Black')
plt.xlabel('Days')
plt.ylabel('No of cakes sold',color='Black')
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.xlim(0,6) #only show x values from 0 to 6
plt.ylim(0,) #only show y values from 0 to 6
plt.grid()
```

```python
# sales data
product=['Laptop','Apple','Samsung','Samsung Fold']
earned=[120000,89000,120000,180000]

plt.barh(product,earned,color='violet',label='Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Earnings')
plt.title('Product Sales')
plt.legend()

plt.show()
```

```python
regions=['North','South','East','West']
revenue=[3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',explode=(0.25,0,0,0))
plt.show()
```

```python
scores=[50,56,89,20,63,45,52,56,45,0,23,78,100]
plt.hist(scores,bins=5,color='red',edgecolor='black')
plt.xlabel('scores')
plt.ylabel('Number of students')
plt.title('Score Distribtion')
plt.grid()
plt.show()
```

```python
hours_studied=[1,2,3,4,5]
marks_scored=[34,78,89,90,100]
marks_pred=[33,70,80,90,99]
plt.scatter(hours_studied,marks_scored,color='red',marker='d',label='Actual Marks')
plt.scatter(hours_studied,marks_pred,color='black',marker='o',label='Marks Predicted')
plt.xlabel('Hours studied')
plt.ylabel('Marks Scored')
plt.legend()
plt.grid()
plt.show()
```
